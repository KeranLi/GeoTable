```python
# import the checked excel file
import pandas as pd

df = pd.read_csv("./1400K.csv", encoding='gb18030')

df
```

```python
# 创建新的DataFrame
new_df = pd.DataFrame()

# 判断Age_238U和Age_206Pb的存在情况，并添加Age列
if 'Published 206Pb/238U age (Ma)' in df.columns and 'Published 206Pb/207Pb age (Ma)' in df.columns:
    mask = (df['Published 206Pb/238U age (Ma)'] > 1200) & (df['Published 206Pb/238U age (Ma)'] > 1200)
    new_df['Age'] = df.loc[mask, 'Published 206Pb/238U age (Ma)']
    new_df.loc[~mask, 'Age'] = df.loc[~mask, 'Published 206Pb/238U age (Ma)']
    new_df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce').dropna().apply(lambda x: round(x, 2))
    new_df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce').dropna().apply(lambda x: round(x, 2))
elif 'Published 206Pb/238U age (Ma)' in df.columns:
    new_df['Age'] = df['Published 206Pb/238U age (Ma)']
    new_df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce').dropna().apply(lambda x: round(x, 2))
    new_df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce').dropna().apply(lambda x: round(x, 2))
elif 'Published 206Pb/238U age (Ma)' in df.columns:
    new_df['Age'] = df['Published 206Pb/238U age (Ma)']
    new_df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce').dropna().apply(lambda x: round(x, 2))
    new_df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce').dropna().apply(lambda x: round(x, 2))
```

```python
# record the latitud and logitud
import numpy as np

lat, lon = np.array(new_df['Latitude']), np.array(new_df['Longitude'])

age = np.array(new_df['Age'])
```

```python
import collections
import re
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# 过滤掉非数值类型的元素，并记录对应的经纬度信息
age_numeric = []
lon_numeric = []
lat_numeric = []
for i in range(len(age)):
    a = age[i]
    match = re.match(r'^(\d+(?:\.\d+)?).*$', str(a))
    if match:
        age_numeric.append(float(match.group(1)))
        lon_numeric.append(lon[i])
        lat_numeric.append(lat[i])

# 定义年龄区间
age_intervals = [
    (0, 500),
    (500, 1000),
    (1000,1500),
    (1500,2000),
    (2000,2500),
    (2500,3000),
    (3000,3500),
    (3500,4000),
    (4000,4500)
    ]

# 初始化区间计数器
age_counts = {interval: 0 for interval in age_intervals}

# 统计不同年龄段的点的数量
for age in age_numeric:
    for interval in age_intervals:
        if interval[0] <= age < interval[1]:
            age_counts[interval] += 1
            break

# 打印统计结果
for interval, count in age_counts.items():
    print(f"Age Interval: {interval}, Count: {count}")
```
